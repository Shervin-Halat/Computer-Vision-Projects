```python
h1 = cv2.imread('Hist1.jpeg',0)
h2 = cv2.imread('Hist2.jpg',0)
h2_color = cv2.imread('Hist2.jpg')
```

```python
hist1 = cv2.calcHist([h1],[0],None,[256],[0,256])
hist2 = cv2.calcHist([h2],[0],None,[256],[0,256])
```

```python
plt.plot(hist1,color = 'deepskyblue')
plt.title('Hist1 histogram',size = 20)
plt.show()
```

```python
plt.plot(hist2,color = 'deepskyblue')
plt.title('Hist2 histogram',size = 20)
plt.show()
```

# Histogram Equalization:

## hist1:

```python
eq1 = cv2.equalizeHist(h1)
eqhist1 = cv2.calcHist([eq1],[0],None,[256],[0,256])

plt.plot(eqhist1,color = 'deepskyblue')
plt.title('Hist1 histogram',size = 20)
plt.show()
plt.imshow(eq1,'gray')
plt.axis('off')
plt.show()
```

## hist2:

```python
h2_chs = cv2.split(h2_color)
eq2_chs = []
for ch in h2_chs:
    eq2_chs.append(cv2.equalizeHist(ch))

eq2 = cv2.merge(eq2_chs)

eq2_gray = cv2.cvtColor(eq2, cv2.COLOR_BGR2GRAY)
eqhist2 = cv2.calcHist([eq2_gray],[0],None,[256],[0,256])
plt.plot(eqhist2,color = 'deepskyblue')
plt.show()

eq2 = cv2.cvtColor(eq2, cv2.COLOR_BGR2RGB)
plt.imshow(eq2)
plt.axis('off')
plt.show()
```

## Final plots:

```python
# firtst image:
plt.subplot(2,2,1)
plt.imshow(h1,'gray',vmin = 0, vmax = 255)
plt.title('original image')
plt.subplot(2,2,2)
plt.plot(hist1,color = 'deepskyblue')
plt.title('original histogram')
plt.subplot(2,2,3)
plt.imshow(eq1,'gray',vmin = 0, vmax = 255)
plt.title('equalized image')
plt.subplot(2,2,4)
plt.plot(eqhist1,color = 'deepskyblue')
plt.title('equalized histogram')
plt.suptitle('Hist1 image',size = 20)
# plt.savefig('1a.png')
plt.show()
```

```python
# second image:
plt.subplot(2,2,1)
h2_color_rgb = cv2.cvtColor(h2_color,cv2.COLOR_BGR2RGB)

plt.imshow(h2_color_rgb,'gray',vmin = 0, vmax = 255)
plt.title('original image')
plt.subplot(2,2,2)
plt.plot(hist2,color = 'deepskyblue')
plt.title('original histogram')
plt.subplot(2,2,3)
plt.imshow(eq2,'gray',vmin = 0, vmax = 255)
plt.title('equalized image')
plt.subplot(2,2,4)
plt.plot(eqhist2,color = 'deepskyblue')
plt.title('equalized histogram')
plt.suptitle('Hist2 image',size = 20)
# plt.savefig('1b.png')
plt.show()
```
