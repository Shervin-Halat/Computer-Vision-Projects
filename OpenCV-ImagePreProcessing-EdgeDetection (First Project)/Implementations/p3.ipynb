```python
def high_boost(image,A):
    blurred = cv2.GaussianBlur(image, (7,7),0)
    sharpened = cv2.addWeighted(image, A+1, blurred, -A, 0)
    sharpened = cv2.cvtColor(sharpened,cv2.COLOR_BGR2RGB)
    return sharpened
```

## Final plots:

```python
a = [2,10,15,25]
for i in range(4):
    plt.subplot(3,2,i+1)
    plt.imshow(high_boost(img,a[i]))
    plt.title('A = {}'.format(a[i]))
    plt.axis('off')
plt.suptitle('Unsharp Masking (High Boost)')
# plt.savefig('3.png')
plt.show()
```
