```python
img = cv2.imread('Shapes.png')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray,'gray')
```

```python
blurred = cv2.GaussianBlur(img_gray,(5,5),0)
thresh = cv2.threshold(blurred,108,255,0)
```

```python
thresh = thresh[1]
```

```python
plt.imshow(thresh,'gray',vmin= 0,vmax = 255)
```

```python
contours, hierarchy = cv2.findContours(thresh,  cv2.RETR_EXTERNAL  , cv2.CHAIN_APPROX_NONE) 
```

```python
%matplotlib inline
```

```python
img = cv2.imread('Shapes.png')
black = np.zeros(img.shape)

cv2.drawContours(black, contours, -1,(0,170,120), 2)
plt.subplot(2,1,1)
plt.imshow(black.astype(np.int),vmin  =0 ,vmax = 255)
plt.title('separated contours',size = 15)
plt.xticks([])
plt.yticks([])

img2 = img.copy()
cv2.drawContours(img2, contours, -1,(0,170,120), 2)
plt.subplot(2,1,2)
plt.imshow(img2)
plt.title('contours on original image',size = 15)
plt.xticks([])
plt.yticks([])

plt.savefig('p3.png')
plt.show()
```
