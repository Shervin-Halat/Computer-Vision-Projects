```python
img = cv2.imread('Circles.png')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray_blurred = cv2.blur(img_gray, (3, 3)) 
```

```python
detected_circles = cv2.HoughCircles(gray_blurred, cv2.HOUGH_GRADIENT, 1, 10\
                                    , param1 = 120, param2 =120, minRadius = 10, maxRadius = 110) 
```

```python
img = cv2.imread('Circles.png')
img2 = img
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

if detected_circles is not None: 
    detected_circles = np.uint16(np.around(detected_circles)) 
    for pt in detected_circles[0, :]: 
        a, b, r = pt[0], pt[1], pt[2] 
        cv2.circle(img2, (a, b), r, (255,0,0), 4) 
        cv2.circle(img2, (a, b), 1, (255, 0, 0), 8)
        
plt.imshow(img2)
plt.xticks([])
plt.yticks([])
plt.title('detected circles by HoughCircles\nnumber of circles = {}'.format(len(detected_circles[0,:])),size = 20)
plt.savefig('p2.png')
plt.show()
```
