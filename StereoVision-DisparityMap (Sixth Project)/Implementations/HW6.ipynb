# sawtooth best model:

```python
#sawtooth parameters:
num_disparities = 16
block_size = 11
min_disparity = 1
texture_threshold = 10
speckle_size = 200
speckle_range = 2
uniqueness_ratio = 40
prefilter_cap = 20
prefilter_size = 5
```

```python
disparity1 = []
for i in range(len(data1)-3):
    imgL, imgR = data1[i+2], data1[i+3]
    stereo = cv2.StereoBM_create(numDisparities=num_disparities, blockSize=block_size)
    stereo.setMinDisparity(min_disparity)
    stereo.setUniquenessRatio(uniqueness_ratio)
    stereo.setTextureThreshold(texture_threshold)
    stereo.setSpeckleRange(speckle_range)
    stereo.setSpeckleWindowSize(speckle_size)
    stereo.setPreFilterCap(prefilter_cap)
    stereo.setPreFilterSize(prefilter_size)
    out = stereo.compute(imgL,imgR)
    out[out==(min_disparity-1)*16] = 0
    out[out<0] = 0
    disparity1.append(out)
```

```python
plt.subplot(1,2,1)
plt.imshow(data1[0],'gray')
plt.subplot(1,2,2)
plt.imshow(disparity1[1],'gray')
plt.show()
```
